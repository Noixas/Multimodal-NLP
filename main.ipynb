```python
default_args = """
                --config config/uniter-base.json
                --data_path ./dataset --model_path ./model_checkpoints 
                --pretrained_model_file uniter-base.pt 
                --feature_path ./dataset/own_features
                --model_save_name meme.pt
                
                
                --scheduler warmup_cosine 
                --warmup_steps 500 
                --max_epoch 30 
                --batch_size 16 
                --patience 5 
                --gradient_accumulation 2
                --lr 3e-5 
                --pos_wt 1
                --normalize_img
                """
"""
Creates the arguments for the run, default arguments represent the baseline.
"""
def return_run_args(seed=43, linear_layers=1, gender_race_hidden_size=0, 
                    upsample_multiplier=0, upsample_options = "", 
                    filter_text=False, note = "Baseline"):
    print("upsample_options",upsample_options)
    run_args = str(default_args + 
                " --seed " + str(seed) + 
                " --linear_layers " + str(linear_layers) + 
                " --gender_race_hidden_size "+str(gender_race_hidden_size) + 
                " --upsample_multiplier " +str(upsample_multiplier) + 
                " --upsample_options " + '"'+str(upsample_options)+'"' + 
                " --note " + '"'+note +'"')
    if filter_text:
       run_args += " --filter_text " 
    return run_args
```

## Replicate Results
Each of the following cells defines the arguments for a modification and proceeds to train the model with those arguments.    
In the paper we show the average of 3 seeds, for demonstrative purposes the cells bellow will run with only one seed (43), feel free to change the seed. Seeds used in the paper are [18, 43, 99]

### Baseline (with normalization, 1 linear layer)

```python
note                     = "Baseline (with normalization, 1 linear layer)"
seed                     = 43
linear_layers            = 1 
gender_race_hidden_size  = 0
upsample_multiplier      = 0
upsample_options         = "0" 
filter_text              = False 
                        
run_args = return_run_args(seed                     = seed, 
                           linear_layers            = linear_layers, 
                           gender_race_hidden_size  = gender_race_hidden_size, 
                           upsample_multiplier      = upsample_multiplier,
                           upsample_options         = upsample_options, 
                           filter_text              = filter_text, 
                           note                     = note)
train_model(run_args)
```
